Escribir una función que devuelva un Hexágono.

```python
def hexagono():
    for i in range(6,14,2):
        print(('*'*i).center(30))
    
    for j in range(10,4,-2):
        print(('*'*j).center(30))
```

c) Generar 3 variables cualitativas ordinales con los siguientes criterios:
    
    ★	Si es mayor a la “media + 1*desviación estándar” → Alto
    ★	Si está entre “media - 1*desviación estándar” y “media + 1*desviación estándar” → Medio
    ★	Si es menor a la “media - 1*desviación estándar” → Bajo

Nota: Mostrar con las nuevas variables generadas. Elegir con libertad las variables a trabajar.


```python
def variable_cualitativa(DataFrame, variable):
    
    variable_array=np.array(variable)
    
    variable_cualitativa=[]
    
    for i in variable_array:
        if i > variable_array.mean()+1*np.std(variable_array):
            variable_cualitativa.append("Alto")
            
        elif (i>variable_array.mean()-1*np.std(variable_array)) and (i< variable_array.mean()+1*np.std(variable_array)):
            variable_cualitativa.append("Medio")
            
        elif i<variable_array.mean()-1*np.std(variable_array):
            variable_cualitativa.append("bajo")
            
    variable_dataframe=pd.DataFrame(variable_cualitativa)
    
    Dataframe_final=pd.concat([DataFrame, variable_dataframe], axis=1)
    
    return Dataframe_final
```

d) Crear una función que muestre los siguientes gráficos para las variables numéricas según una variable numérica:
    
    ❖	Cajas y bigotes
    
    ❖	Barras de la media
    
    ❖	Histograma con Curva


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plots(DataFrame):
    
    #columnas númericas
    columnas = DataFrame.select_dtypes(include=["float64", "int64"]).columns
    
    #gráficas de Cajas y bigotes
    
    sns.boxplot(data=DataFrame[columnas], palette="rainbow", orient="h")
    plt.show()
    
    #gráfica Barras de la media
    
    DataFrame.mean().plot(kind="bar")
    plt.show()
    
    #Histograma con Curva
    
    for i in columnas:
        sns.set(color_codes=True)
        sns.distplot(DataFrame["{}".format(i)])
        plt.show()
```

```python
plots(df2)
```

e) Crear una función que devuelva un mapa de calor con la correlación de todas las variables numéricas (incluye las generadas), para cada país.

```python
def mapa_calor_pais(Dataframe):
    list_pais =  Dataframe["Pais_Origen"].unique()
    
    Dataframe_index_pais = Dataframe.set_index("Pais_Origen")
    
    columnas= Dataframe_index_pais.select_dtypes(include=["float64", "int64"]).columns
    
    Dataframe_index_pais = Dataframe_index_pais[columnas]
    
    
    for i in list_pais:
        
        df=Dataframe_index_pais.loc["{}".format(i), :]
        sns.heatmap(df.corr())
        print( "######################## PAIS {}  ###################".format(i))
        plt.show()
```

```python
mapa_calor_pais(df)
```
